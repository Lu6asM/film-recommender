# **Documentation**

`name.basics.tsv.gz`

*   **nconst** *(string)* - alphanumeric unique identifier of the name/person
*   **primaryName** *(string)*– name by which the person is most often credited
*   **birthYear** *(YYYY)* – in YYYY format
*   **deathYear** *(YYYY)* – in YYYY format if applicable, else '\N'
*   **primaryProfession** *(array of strings)* – the top-3 professions of the person
*   **knownForTitles** *(array of tconsts)* – titles the person is known for


**Utilité** : Répertorie les personnes (réalisateurs, scénaristes, acteurs, etc.) et des informations sur elles, telles que leur date de naissance et leurs œuvres connues.

**Application** : Permet des analyses sur la carrière des individus, comme l'identification des réalisateurs les plus influents ou des acteurs ayant le plus de films à leur actif.

```python
name_basics = pd.read_csv('https://datasets.imdbws.com/name.basics.tsv.gz', sep = '\t', nrows=1000)
name_basics
```

`title.akas.tsv.gz`

*   **titleId** *(string)* - a tconst, an alphanumeric unique identifier of the title
*   **ordering** *(integer)* – a number to uniquely identify rows for a given titleId
*   **title** *(string)* – the localized title
*   **region** *(string)* - the region for this version of the title
*   **language** *(string)* - the language of the title
*   **types** *(array)* - Enumerated set of attributes for this alternative title.  One or more of the following: "alternative", "dvd", "festival", "tv",  "video", "working", "original", "imdbDisplay". New values may be added in the future without warning
*   **attributes** *(array)* - Additional terms to describe this alternative title, not enumerated
*   **isOriginalTitle** *(boolean)* – 0: not original title; 1: original title

**Utilité** : Contient des titres alternatifs et des localisations spécifiques des titres de films/séries.

**Application** : Permet de voir comment un même film ou une série est connu sous différents titres dans le monde (par exemple, les traductions et adaptations locales).

```python
title_akas = pd.read_csv('https://datasets.imdbws.com/title.akas.tsv.gz', sep = '\t', nrows=1000)
title_akas
```

`title.basics.tsv.gz`

*   **tconst** *(string)* - alphanumeric unique identifier of the title
*   **titleType** *(string)* – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
*   **primaryTitle** *(string)* – the more popular title / the title used by the filmmakers on promotional materials at the point of release
*   **originalTitle** *(string)* - original title, in the original language
*   **isAdult** *(boolean)* - 0: non-adult title; 1: adult title
*   **startYear** *(YYYY)* – represents the release year of a title. In the case of TV Series, it is the series start year
*   **endYear** *(YYYY)* – TV Series end year. '\N' for all other title types
runtimeMinutes – primary runtime of the title, in minutes
*   **genres** *(string array)* – includes up to three genres associated with the title

**Utilité** : Représente les données de base sur chaque titre (films, séries TV, épisodes, etc.).

**Application** : Utilisé pour obtenir des informations essentielles comme le type de contenu (film, série, etc.), le titre original, les genres et la durée. C’est crucial pour des analyses comme le nombre de films sortis par année, les genres les plus populaires, etc.

```python
title_basics = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep = '\t', nrows=1000)
title_basics
```

`title.crew.tsv.gz`

*   **tconst** *(string)* - alphanumeric unique identifier of the title
*   **directors** *(array of nconsts)* - director(s) of the given title
*   **writers** *(array of nconsts)* – writer(s) of the given title

**Utilité** : Liste les réalisateurs et les scénaristes associés à chaque titre.

**Application** : Permet d'identifier qui a travaillé sur quel projet, ce qui est important pour des analyses sur les collaborations récurrentes entre réalisateurs et scénaristes ou l'impact d'un créateur sur le succès d'un projet.

```python
title_crew = pd.read_csv('https://datasets.imdbws.com/title.crew.tsv.gz', sep = '\t', nrows=1000)
title_crew
```

`title.episode.tsv.gz`

*   **tconst** *(string)* - alphanumeric identifier of episode
*   **parentTconst** *(string)* - alphanumeric identifier of the parent TV Series
*   **seasonNumber** *(integer)* – season number the episode belongs to
*   **episodeNumber** *(integer)* – episode number of the tconst in the TV series

**Utilité** : Fournit des informations spécifiques aux épisodes de séries télévisées.

**Application** : Utile pour analyser la structure des séries (nombre de saisons, nombre d'épisodes par saison, etc.).

```python
title_episode = pd.read_csv('https://datasets.imdbws.com/title.episode.tsv.gz', sep = '\t', nrows=1000)
title_episode
```

`title.principals.tsv.gz`

*   **tconst** *(string)* - alphanumeric unique identifier of the title
*   **ordering** *(integer)* – a number to uniquely identify rows for a given titleId
*   **nconst** *(string)* - alphanumeric unique identifier of the name/person
*   **category** *(string)* - the category of job that person was in
*   **job** *(string)* - the specific job title if applicable, else '\N'
*   **characters** *(string)* - the name of the character played if applicable, else '\N'

**Utilité** : Contient des informations sur les membres principaux d'un titre, tels que les acteurs et autres membres de l'équipe.

**Application** : Utile pour analyser la distribution des rôles, repérer les acteurs les plus prolifiques ou comprendre la répartition des rôles dans les productions.

```python
title_principals = pd.read_csv('https://datasets.imdbws.com/title.principals.tsv.gz', sep = '\t', nrows=1000)
title_principals
```

`title.ratings.tsv.gz`

*   **tconst** *(string)* - alphanumeric unique identifier of the title

**Utilité** : Contient les notes moyennes et le nombre de votes des utilisateurs pour chaque titre.

**Application** : Permet des analyses sur la popularité et la qualité perçue des films et séries, comme la comparaison des films les mieux notés par genre ou par année.

```python
title_rating = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep = '\t', nrows=1000)
title_rating
```
