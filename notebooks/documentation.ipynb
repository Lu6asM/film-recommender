# **Nettoyage**

```python

#! === SETUP ===
import pandas as pd

df_title_basics = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep = '\t', compression='gzip', na_values='\\N')
df_title_ratings = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep = '\t', compression='gzip', na_values='\\N')
```

```python

#! === CLEANUP ===
#? title_basics

#* Garder uniquement les films
df_title_basics = df_title_basics[df_title_basics['titleType'] == 'movie']

#* Supprimer les lignes avec des donnes manquantes essentielles
df_title_basics = df_title_basics.dropna(subset=['startYear', 'genres'])

#* Convertir startYear en entier et filtrer les annes valides
df_title_basics['startYear'] = df_title_basics['startYear'].astype(int)
df_title_basics = df_title_basics[(df_title_basics['startYear'] >= 1950) & (df_title_basics['startYear'] <= 2023)]


#! === CLEANUP ===
#? title_rating

#* Supprimer les films avec peu de votes
df_title_ratings = df_title_ratings[df_title_ratings['numVotes'] > 500]
```

# **Fusion**

```python

#! === MERGEUP ===
df_merged = df_title_basics.merge(df_title_ratings, on="tconst")
```

```python

#! === CLEANUP ===
df_merged['Dcennie'] = (df_merged['startYear'].astype(int) // 10) * 10

df_merged = df_merged[['tconst', 'primaryTitle', 'startYear', 'genres', 'averageRating', 'numVotes', 'Dcennie']]

df_merged = df_merged.rename(columns={'tconst': 'ID',
                                      'primaryTitle': 'Titre',
                                      'startYear': 'Anne',
                                      'genres':  'Genres',
                                      'averageRating':  'Note_Moyenne',
                                      'numVotes': 'Votes'
                                      })

df_merged
```

```python

#! === EXPORT ===
df_merged.to_csv("df_merged_cleaned.csv", index=False)
```

```python
def categorize_votes(votes):
    if votes < 10000:
        return "Peu connu"
    elif 10000 <= votes < 50000:
        return "Connu"
    elif 50000 <= votes < 100000:
        return "Populaire"
    else:
        return "Trs populaire"

df['Popularity'] = df['Votes'].apply(categorize_votes)
```
