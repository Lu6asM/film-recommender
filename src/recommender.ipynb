# **Importation**

```python
import pandas as pd

df_movie = pd.read_csv("../data/processed/df_movie_cleaned.csv")

df_movie
```

# **Système de recommandation**

```python
def recommend_by_genre(df, genre, top_n=5):
    filtered = df[df['Genres'].str.contains(genre, na=False)].sort_values(by='Note', ascending=False)
    return filtered[['Titre', 'Titre Français', 'Année', 'Note']].head(top_n)

print(recommend_by_genre(df_movie, 'Action'))
```

```python
print(recommend_by_genre(df_movie, 'Comedy'))
```

```python
print(recommend_by_genre(df_movie, 'Drama'))
```
